```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Exemplo fictcio de dados
dados = {
    "mes": [1, 2, 3, 4, 5, 6, 7],
    "faturamento": [6317786.28, 6468952.36, 6577028.23, 6527821.45, 6674084.12, 6758464.07, 6882779.07]
}
df = pd.DataFrame(dados)

# Preparar dados
X = df["mes"].values.reshape(-1,1)
y = df["faturamento"].values

# Criar modelo
modelo = LinearRegression()
modelo.fit(X, y)

# Prever prximos meses
proximos_meses = np.array([8, 9, 10]).reshape(-1,1)
previsao = modelo.predict(proximos_meses)

# Plot
plt.figure(figsize=(8,5))
plt.scatter(X, y, color="blue", label="Histrico")
plt.plot(X, modelo.predict(X), color="red", label="Tendncia Linear")
plt.scatter(proximos_meses, previsao, color="green", label="Previso")

# Adicionar rtulos em milhes
for i, valor in enumerate(y):
    plt.text(X[i], valor, f"R$ {valor/1e6:.2f}M", fontsize=9, ha='center', va='bottom')

for i, valor in enumerate(previsao):
    plt.text(proximos_meses[i], valor, f"R$ {valor/1e6:.2f}M", fontsize=9, ha='center', va='bottom', color="green")

plt.xlabel("Ms")
plt.ylabel("Faturamento (R$)")
plt.title("Previso Linear de Faturamento")
plt.legend()
plt.tight_layout()
plt.show()
```
